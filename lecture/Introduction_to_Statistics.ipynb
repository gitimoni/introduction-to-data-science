1.   **What is Statistics?**

Statistics is the study of how best to collect, analyze, and draw conclusions from data - observations collected from the likes of field notes, surveys and experiments. It is the art of converting data into useful information.

Examples of Applications of Statistics:

*   Predictive Modelling: Using monthly sales figures in a company and analysing it to predict sales for next month 
*   Clinical Trials: A clinical trial involves selecting a population/sample size, defining the time range over which to monitor the treatment, designing the phases, and selecting parameters that will help decide how effective the treatment is and if it is better than an existing one
*   Research:  Researchers test hypotheses, confirm (or reject) theories, and arrive at reliable conclusions

**Key Statiscal Concepts**



1.   **Population**

A population is the set of all items of interest in a statistical problem. It is frequently very large; may in fact, be infinitely large and does not necessarily refer to a group of people. 

For example all the car sales in Toronto, all the students enrolled in a program.

Some measurements of the population may be of interest to us: 
average, frequency and variation. A descriptive measure of a population is called a **parameter**.

2.   **Sample**

A sample is a subset of data drawn from the population using various sampling strategies. The data we collect for our study in almost all cases are sample data. Descriptive measures of a sample are called **statistics**. We use **statistics** to make estimation about **parameters**. 


3.  **Descriptive Statistics**

Descriptive statistics are quantitave or graphical methods used to describe or summarize a collection of data. Descriptive statistics include the following summaries:

*   Measures of Central Tendency: They represent the center point or typical value of a dataset. Mean (average value) and Median (middle value) are popular choices of measurement for numerical data where Mode (most frequent value) can be used for categorical data.

*   Measures of Variability: Variability describes how far apart data points lie from each other and from the center of numerical data. Common metrics are range, interquartile range, standard deviation and variance.

* Measures of Relative Standing: These show the location of data values relative to the other values within a data set. Measures include quartiles, percentiles and other quantiles, z-scores, etc.

* Measures of Relationship: They show a relationship between two or more variables or two or more sets of data. Eg: Pearson Correlation, Spearman Correlation, etc.

* Graphical Representations: Visualising data is one of the most intuitive method to understand data better. Line plots, Scatter plots, Bar plots, histograms, pie charts, etc. are some of the many graphical representations used in data analysis. We will explore these using **Matplotlib**, a powerful visualisation tool available in Python.


4. **Inferential Statistics:** 

Inferential Statistics is the process of making an estimate, prediction, or decision about a population based on sample data. Because population are almost always very large, investigating each member of the population would be impractical and expensive. It is far easier and cheaper to take a sample from the population of interest and obtain conclusions or make estimates about the population on the basis of information provided by the sample.

However, such conclusions and estimates are not always going to be correct for various reasons. The results are very much dependent on your samples. This is why much thought is given to sampling strategies in a goal to obtain a sample that is representative of the population data. 

To measure how good an estimate is, we incorporate probability as a measure of reliability. Eg: Confidence interval estimates of population parameter, Hypothesis testing, etc.
